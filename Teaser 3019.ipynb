### William's Prime

William was searching for a number he could call his own. By consistently replacing digits with letters, he found a number represented by his name **WILLIAM**.

He noticed that he could break **WILLIAM** down into three smaller numbers represented by **WILL, I** and **AM**, where **WILL** and **AM** are prime numbers.

He then calculated the product of the three numbers **WILL, I** and **AM**.

If I told you how many digits there are in the product, you would be able to determine the number represented by **WILLIAM**.

First let's make a tables of two and four digit primes

```python
def is_prime(int):
    for i in range(2, round((int)**(1/2))+1):
        prm = True
        if int % i == 0:
            prm = False
            break
    return prm

td_primes = []
for i in range(10,100):
    if is_prime(i) == True:
        td_primes.append(i)
print(td_primes)


fd_primes = []
for i in range(1000,10000):
    if is_prime(i) == True:
        fd_primes.append(i)
print(fd_primes)
```

Now we convert them to strings and see what options we have for WILLIAM given WILL and AM are prime (i.e. the constraints are that the last two digits of the four digit prime are the same as each other and different to the other two which are also different from each other...).

```python
wills = []
for num in fd_primes:
    num = str(num)
    if num[2] == num[3] and num[0] != num[1] and num[0] != num[2] and num[1] != num[2]:            
        wills.append(num)
print(wills)
```

```python
ams = []
for num in td_primes:
    num = str(num)
    if num[0] != num[1]:           
        ams.append(num)

willams = []        
for will in wills:
    for am in ams:
        letts = []
        for wilet in will:
            letts.append(wilet)
        for amlet in am:
            letts.append(amlet)
        if len(letts) - len(set(letts)) == 1:
            willams.append([will, am])

for row in willams:
    i = row[0][1]
    row.append(i)
print(willams)
            
```

Now we have all the possibilities we need to see how long the product is and see which length is unique

```python
prods = []
digits = {}
digits_prods = {}
for row in willams:
    prod = 1
    for num in row:
        intnum = int(num)
        prod *= intnum   
    prods.append(prod)
    prodlen = str(len(str(prod)))
    if prodlen not in digits:
        digits[prodlen] = 1
        digits_prods[prodlen] = []
        digits_prods[prodlen] = [row]
    else:
        digits[prodlen] += 1
        digits_prods[prodlen].append(row)
print(digits)
for key in digits:
    if digits[key] == 1:
        answer = digits_prods[key]
print("WILLIAM =", answer[0][0]+answer[0][2]+answer[0][1])
```
